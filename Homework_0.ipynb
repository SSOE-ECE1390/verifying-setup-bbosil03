```python
#Import all of the required libraries
import cv2

import os
import numpy as np

import matplotlib.pyplot as plt

#Read in the image of the rocket and display it
apollo_img = cv2.imread(os.path.relpath("Data/Apollo_11_Launch.jpg"), 0) #read as grayscale

plt.imshow(apollo_img, cmap="gray")
plt.title("Apollo 11 Launch")
```

```python
#Crop out just the rocket
cropped_apollo = apollo_img[90:650, 450:750]
plt.imshow(cropped_apollo, cmap="gray")
plt.title("Cropped image")
```

```python
#Brighten the image by a a value of 50
brightened_image = apollo_img + 50
plt.imshow(brightened_image, cmap="gray")
plt.title("Brightened image (+ 50)")
#The resulting image has some overflow and part of the image becomes darker
```

```python
#Create one copy of the image rotated to the left and one rotated right
rotated_left = cv2.rotate(apollo_img, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotated_right = cv2.rotate(apollo_img, cv2.ROTATE_90_CLOCKWISE)

#Plot the images in a 1x2 subplot
plt.subplot(121); plt.imshow(rotated_left, cmap="gray"); plt.title("Rotated left")
plt.subplot(122); plt.imshow(rotated_right, cmap="gray"); plt.title("Rotated right")
```

```python
#Saving two different instances of the image
cv2.imwrite("apollo11_rotated_left.png", rotated_left)
cv2.imwrite("apollo11_brightened.png", brightened_image)
```
